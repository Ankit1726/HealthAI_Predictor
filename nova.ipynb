#### NoveGen Research Lab Project
- Aim: To Identify using model is healthy or unhealthy

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, recall_score

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
```

```python
## Creating Base Line Model
df = pd.read_csv("novagen_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.shape
```

```python
## Feature Spllit & Target
x = df.drop("Target",axis=1)
y = df['Target']
```

```python
x.head()
```

```python
y.head()
```

```python
## Train Test Split
X_train,X_test,Y_train,Y_test = train_test_split(
     x,y,test_size=0.3,random_state=42
)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

- Linear Regression

```python
## Logistic REgression With Regularization
log_reg = LogisticRegression(
     penalty="l2",
     solver="liblinear",
     max_iter=1100
)
log_reg.fit(X_train_scaled, Y_train)
```

```python
y_pred = log_reg.predict(X_test_scaled)
```

```python
print(f"Logisitic Regression Accuracy:{accuracy_score(Y_test,y_pred)*100}%")
print(f"Logisitic Regression Recall:{recall_score(Y_test,y_pred)*100}%")
print(f"{classification_report(Y_test,y_pred)}")
```

- KNN

```python
knn = KNeighborsClassifier(
    n_neighbors=5,
    metric="euclidean"
)

knn.fit(X_train_scaled,Y_train)
```

```python
y_pred = knn.predict(X_test_scaled)
```

```python
print(f"KNN Accuracy:{accuracy_score(Y_test,y_pred)*100}%")
print(f"KNN Recall:{recall_score(Y_test,y_pred)*100}%")
print(f"{classification_report(Y_test,y_pred)}")
```

- Random Forest(Ensemble Learning)

```python
rf = RandomForestClassifier(
    n_estimators=201,
    max_depth=None,
    random_state=42
)
rf.fit(X_train,Y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print(f"Random Forest Accuracy: {accuracy_score(Y_test,y_pred)*100}%")
print(f"Random Forest Recall: {recall_score(Y_test,y_pred)*100}%")
print(f"{classification_report(Y_test,y_pred)}")
```

- Gradient Boosting (Ensemble Learning)

```python
gb = GradientBoostingClassifier(
    n_estimators=151,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
gb.fit(X_train,Y_train)
```

```python
y_pred = gb.predict(X_test)
```

```python
print(f"GB Accuracy: {accuracy_score(Y_test,y_pred)*100}%")
print(f"GB Recall: {recall_score(Y_test,y_pred)*100}%")
print(f"{classification_report(Y_test,y_pred)}")
```

- Voting Classifier(Ensemble Learning)

```python
voting_clf = VotingClassifier(
    estimators=[
        ("lr", LogisticRegression(max_iter=1000, solver="liblinear")),
        ("knn", KNeighborsClassifier(n_neighbors=5)),
        ("rf", RandomForestClassifier(n_estimators=201, random_state=42))
    ],
    voting="soft"
)
voting_clf.fit(X_train_scaled,Y_train)
```

```python
y_pred = voting_clf.predict(X_test_scaled)
```

```python
print(f"VC accuracy: {accuracy_score(Y_test,y_pred)*100}%")
print(f"VC recall: {recall_score(Y_test,y_pred)*100}%")
print(f"{classification_report(Y_test,y_pred)}")
```

#### Final Results
